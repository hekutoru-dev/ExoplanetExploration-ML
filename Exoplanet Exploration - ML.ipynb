# General Dependencies

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("cumulative.csv")
df.columns
```

```python
df = df.drop(["rowid", "kepid", "kepoi_name", "kepler_name", "koi_pdisposition", "koi_score", "koi_tce_delivname"], axis=1)
```

```python
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.columns
df.head()
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
X = df.drop('koi_disposition', axis=1)
y = df['koi_disposition']
```

```python
# Explore X
X.head()
```

```python
df['koi_disposition'].value_counts()
```

```python
# Explore y
print (y.unique())
y.head()
```

```python
# Split data into training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)
```

```python
# Scale training and test data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

# Train the Support Vector Machine

```python
# Fit to the training data and validate with the test data
from sklearn.svm import SVC 
model = SVC() # default is RBF, as it's multiclass
model.fit(X_train_scaled, y_train) # from preprocessing.
#predictions = model.predict(X_test)
```

---

```python
# Test many gamma and C params.
gamma = [1, 5, 10, 20, 30, 40, 50, 60]
C = [0.01, 0.1, 1, 2, 3, 4, 5]

for this_gamma in gamma:
    for this_C in C:
        clf = SVC(kernel = 'rbf', gamma = this_gamma, C = this_C).fit(X_train_scaled, y_train)
        print(f"Training Data Score with Gamma {this_gamma} and C {this_C}: {clf.score(X_train_scaled, y_train)}")
        print(f"Testing Data Score: with Gamma {this_gamma} and C {this_C}: {clf.score(X_test_scaled, y_test)}")
```

```python
# from
# gamma = [0.01, 0.1, 1, 5, 10, 50]
# C = [0.0001, 0.0005, 0.001, 0.005, 0.01, 0.1, 1, 10]
this_gamma = 1
this_C = 40
clf = SVC(kernel = 'rbf', gamma = this_gamma, C = this_C).fit(X_train_scaled, y_train)
```

```python
print(f"Training Data Score: {clf.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {clf.score(X_test_scaled, y_test)}")
```

```python
from sklearn.metrics import confusion_matrix
predictions = clf.predict(X_test_scaled)
confusion = confusion_matrix(y_test, predictions)
confusion
```

```python
from sklearn.metrics import classification_report
#predictions = clf.predict(X_test)
print(classification_report(y_test, predictions))
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the `C` and `gamma` parameters

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [30, 32, 34, 36, 38, 40, 42, 44, 46, 48],
              'gamma': [1, 3, 5, 7, 10]}
grid = GridSearchCV(clf, param_grid, verbose=3)
```

```python
# Train the model with GridSearch
# Fit the model using the grid search estimator. 
# This will take the SVC model and try each combination of parameters
grid.fit(X_train_scaled, y_train)
```

```python
# Listing the best parameters for this dataset & best score.
print(grid.best_params_)
print(grid.best_score_)
```

```python
# Make predictions with the hypertuned model
predictions = grid.predict(X_test_scaled)
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```

```python
from sklearn.metrics import confusion_matrix
predictionsgrid = grid.predict(X_test_scaled)
confusion = confusion_matrix(y_test, predictionsgrid)
confusion
```
